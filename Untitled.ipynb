#### Sales Analysis

```python
import pandas as pd
import os
```

#### Merging 12 months to a single file

```python
# To view all files in folder
files = [file for file in os.listdir('./SalesAnalysis/Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./SalesAnalysis/Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

#### Read data to dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

#### Cleaning up data

```python
# drop rows of NaN

nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### find and delete 'Or'

```python

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

#### convert columns to correct type

```python

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # to int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # to int

all_data.head()
```

#### augment data with additional columns


```python
# Add Month Column
# use the first two column in the order_date column convert to string

all_data['Month'] = all_data["Order Date"].str[0:2]
all_data['Month'] = all_data["Month"].astype('int32')

all_data.head()
```

```python
# Added Sales Column
```

```python
# created a sales and month column
```

#### Best month for sales and how much money was earned 

```python
# to get sales = Qunatity Oredered x Price Each

all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
# To get the best month of sales

results = all_data.groupby('Month').sum()
```

```python
# To plot the result

import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in usd ($)')
plt.xlabel('Month number')
plt.show()
```

#### What City had the highest number of Sales

#### add a city column

```python
# Split the 'Purchase Address' to get the cities using .apply   method
# lambda 'x' - reps the cell content
# lambda is used to grab cell content

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ') ')

#all_data = all_data.drop(columns='Column', inplace=True)

all_data.head()
```

```python
# To get the city with highest sales

results = all_data.groupby('City').sum()
results
```

```python
# To plot the result

import matplotlib.pyplot as plt

cities = [city for city , df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in usd ($)')
plt.xlabel('City name')
plt.show()
```

#### Products most often sold together 

```python
# they have the same order id
# count all duplicate order id together

all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
#all_data.head(20)

# group order_id by columns
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
# To count the pairs of occurence('same order id') 
# counting unique pairs of number into a python dictionary

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(20):
    print(key, value)
```

#### Products sold the most and Why

```python
# sum up quantity ordered based on grouping by the products 

product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products,  quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```

```python
# Plot two y axis graphs
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
